```python
import os

os.makedirs("docs", exist_ok=True)

files = {
    "docs/index.md": "# 欢迎来到 Jenny 的主页\n\n这是首页。\n\n- [Page 1](page1.md)\n- [Page 2](page2.md)\n",
    "docs/page1.md": "# Page 1\n\n内容。\n\n- [返回首页](index.md)\n- [去 Page 2](page2.md)\n",
    "docs/page2.md": "# Page 2\n\n内容。\n\n- [返回首页](index.md)\n- [去 Page 1](page1.md)\n",
}

for name, content in files.items():
    with open(name, "w", encoding="utf-8") as f:
        f.write(content)

print("docs/ 创建完毕。")
```

```python
mkdocs_yml = """site_name: Jenny Website
nav:
  - Home: index.md
  - Page 1: page1.md
  - Page 2: page2.md
theme:
  name: material
"""

with open("mkdocs.yml", "w", encoding="utf-8") as f:
    f.write(mkdocs_yml)

print("mkdocs.yml 已生成。")
```

```python
import sys, subprocess

result = subprocess.run(
    [sys.executable, "-m", "mkdocs", "build"],
    capture_output=True,
    text=True
)

print("stdout:\n", result.stdout)
print("stderr:\n", result.stderr)
```

```python
import os
import shutil

src = "site"
dst = "."

for item in os.listdir(src):
    s = os.path.join(src, item)
    d = os.path.join(dst, item)

    # 删除旧文件/文件夹
    if os.path.isdir(d):
        shutil.rmtree(d)
    elif os.path.exists(d):
        os.remove(d)

    # 复制新的
    if os.path.isdir(s):
        shutil.copytree(s, d)
    else:
        shutil.copy2(s, d)

print("网站文件已复制到根目录。")
```
